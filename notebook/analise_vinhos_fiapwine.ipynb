```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar dataset
df = pd.read_csv('winequality-red.csv', sep=';')

# Visualizar primeiras linhas
df.head()

# Estrutura do dataset
df.info()

# Estatísticas básicas
df.describe()

# Verificar tamanho
df.shape
```

```python
# Verificar duplicados
df.duplicated().sum()
```

```python
# Remover duplicados (boa prática)
df = df.drop_duplicates()
```

```python
# Verificar valores nulos
df.isnull().sum()
```

🔹 a) Acidez volátil por qualidade

```python
plt.figure()
df.boxplot(column='volatile acidity', by='quality')
plt.title('Acidez Volátil por Qualidade')
plt.suptitle('')
plt.xlabel('Qualidade')
plt.ylabel('Acidez Volátil')
plt.show()
```

🔹 b) Álcool por qualidade

```python
plt.figure()
df.boxplot(column='alcohol', by='quality')
plt.title('Teor Alcoólico por Qualidade')
plt.suptitle('')
plt.xlabel('Qualidade')
plt.ylabel('Álcool')
plt.show()
```

🔹 c) Acidez vs Álcool



```python
plt.figure()
plt.scatter(df['volatile acidity'], df['alcohol'])
plt.xlabel('Acidez Volátil')
plt.ylabel('Álcool')
plt.title('Relação entre Acidez e Álcool')
plt.show()
```

🔹 d) Álcool vs Qualidade

```python
plt.figure()
plt.scatter(df['alcohol'], df['quality'])
plt.xlabel('Álcool')
plt.ylabel('Qualidade')
plt.title('Álcool vs Qualidade')
plt.show()
```
